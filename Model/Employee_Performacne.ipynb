```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Preprocessing: Define feature columns and target column
feature_columns = ['Age', 'Department', 'DailyRate']
target_column = 'Attrition'
```

```python
# Handling missing values and encoding categorical features in a pipeline
preprocessing_pipeline = ColumnTransformer(
    transformers=[
        ('num', SimpleImputer(strategy='mean'), ['Age', 'DailyRate']),
        ('cat', OneHotEncoder(), ['Department'])
    ],
    remainder='passthrough'  # This ensures any additional columns are passed through as-is
)

# Create the ML pipeline with preprocessing and logistic regression model
pipeline = Pipeline([
    ('preprocessing', preprocessing_pipeline),
    ('scaling', StandardScaler()),  # Standardize numerical features
    ('classifier', LogisticRegression())  # You can change this to other classifiers like Decision Tree, Random Forest, etc.
])
```

```python
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
```

```python
# Train the pipeline on the training data
X_train = train_df[feature_columns]
y_train = train_df[target_column]
```

```python
pipeline.fit(X_train, y_train)
```

```python
pipeline.score(X_train,y_train)
```

```python
# Test the pipeline on the testing data
X_test = test_df[feature_columns]
y_test = test_df[target_column]
```

```python
# Make predictions
y_pred = pipeline.predict(X_test)
```

```python
# Evaluate the model's performance
accuracy = accuracy_score(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

print("Model Accuracy:", accuracy)
print("Classification Report:")
print(classification_rep)
```
